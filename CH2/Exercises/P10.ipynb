A row represents a paricular area in Boston like a district. The columns are indicators for that district.

```python
import pandas as pd
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15,15));

pd.plotting.scatter_matrix(boston, ax=ax);
plt.show()
```

# Findings

1. dis appears to be negatively correlated (almost exponentially) with nox.
2. rm is positively correlated with medv - not surprising (more rooms = higher home value).
3. lstat is negatively correlated with rm.
4. age/nox plot has a logarithmic shape 

# Correlated with Crime Rate
